```python
# 🏏 IPL Data Analysis - Google Colab Version

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset (uploaded in /content/)
df = pd.read_csv("/content/matches.csv")
print("✅ Dataset loaded successfully.")
df.head()
```

```python
# Basic info
print("🗂️ Columns:", df.columns.tolist())
print("\n🔎 Dataset shape:", df.shape)
print("\n📊 Data Types:\n", df.dtypes)
print("\n🧼 Missing Values:\n", df.isnull().sum())
```

```python


# Count of match wins
team_wins = df['Winning_Team'].value_counts().head(10)

# Plot
team_wins.plot(kind='bar', color='skyblue')
plt.title("Top 10 Most Successful IPL Teams")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
venue_counts = df['Venue'].value_counts().head(10)

# Plot
venue_counts.plot(kind='barh', color='green')
plt.title("Top 10 IPL Venues")
plt.xlabel("Matches")
plt.ylabel("Venue")
plt.tight_layout()
plt.show()


```

```python
print(df.columns.tolist())
```

```python
# Group by session
season_counts = df['Session'].value_counts().sort_index()

# Plot
season_counts.plot(kind='bar', color='orange')
plt.title("Number of Matches Per Season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Combine both team columns
teams = pd.concat([df['First_Team'], df['Second_Team']])

# Count appearances
team_appearances = teams.value_counts().head(10)

# Plot
team_appearances.plot(kind='bar', color='purple')
plt.title("Top 10 Most Frequent IPL Teams")
plt.ylabel("Match Appearances")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
